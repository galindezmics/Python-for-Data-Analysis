## **Introduction to Data Visualization**

`Data visualization` is the process of presenting information and data in a graphical format. Utilizing visual elements such as charts, graphs, and maps, data visualization tools offer an intuitive approach to interpreting trends, outliers, and patterns within data.

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt
import pandas as pd

# Created a dataframe with sales data
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "March", "Apr", "May"],
    "Sales": [1200, 1500, 1700, 1800, 1600]
})
df

# Plot the data points for (Month, Sales)
plt.plot(df["Month"], df["Sales"], marker='o')

# Set the title of the plot
plt.title("Monthly Sales Data")

# Label the x-axis
plt.xlabel("Month")

# Label the y-axis
plt.ylabel("Sales ($)")

# Display the plot
plt.show()
```

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt
import pandas as pd


# Created a dataframe with sales data
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "March", "Apr", "May"],
    "Sales": [1200, 1500, 1700, 1800, 1600]
})
df

# Generating a bar chart
plt.bar(df["Month"], df["Sales"], color="violet")

# Set the title of the plot
plt.title("Monthly Sales Data")

# Label the x-axis
plt.xlabel("Month")

# Label the y-axis
plt.ylabel("Sales ($)")

# Display the plot
plt.show()
```

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt
import pandas as pd

# Created a dataframe with sales data
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "March", "Apr", "May"],
    "Sales": [1200, 1500, 1700, 1800, 1600]
})
df

# Generating a pie chart
plt.pie(df["Sales"], labels=df["Month"], colors=['tomato', 'cornflowerblue', '#77BFE2', 'blue', 'crimson'], autopct="%1.1f%%")

# Set the title of the plot
plt.title('Monthly Sales Distribution')

# Display the plot
plt.show
```
